```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv("HR_comma_sep.csv")
```

```python
df.groupby('left').mean()
```

```python
subdf = df[['satisfaction_level','average_montly_hours','salary']]
subdf
```

```python
dummy_salary=pd.get_dummies(subdf.salary, prefix="salary") #to get dummy values for words 
dummy_salary
df_with_dummies = pd.concat([subdf,dummy_salary],axis='columns') #to merge both data sets  
df_with_dummies
```

```python
df_with_dummies.drop('salary',axis='columns',inplace=True)
```

```python
df_with_dummies
```

```python
x=df_with_dummies
y=df.left
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.5)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
x_test.head()
```

```python
print(model.predict(x_test.head()))
```

```python
model.score(x_test,y_test)
```
